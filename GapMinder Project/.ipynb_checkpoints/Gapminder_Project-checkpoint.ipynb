## Gapminder Project 

<b> The Project is aimed at visualising the Gapminder dataset. The gapminder data contains data on life expectancy and fertility data by country and year. The Objective is to build an animated scatterplot.<b>

## Read and Preprocess the data

```python
#Import Library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read the datasets into dataframes
df_fer = pd.read_csv('gapminder_total_fertility.csv')
df_pop = pd.read_excel('gapminder_population.xlsx')
df_life = pd.read_excel('gapminder_lifeexpectancy.xlsx')
df_con = pd.read_csv('continents.csv',sep=';')
```

```python
#Clean and tidy the dataframes
df_fer.rename({'Total fertility rate':'Country'}, axis=1, inplace = True)
df_fer.head(3)
```

<b> The data is in the wide format and needs to be converted into the long format </b>

```python
#Convert from Wide to Long Format
df_fer = df_fer.melt(id_vars="Country", var_name="Year",  value_name="Fertility" )
df_fer.head()
```

```python
#Convert the Dataframe Population from Wide to long format
df_pop.rename({'Total population':'Country'}, axis=1, inplace = True)
df_pop = df_pop.melt(id_vars="Country",  var_name="Year",  value_name="population" )
df_pop.head()
```

```python
#Convert the Dataframe Life Expectancy from Wide to long format
df_life.rename({'Life expectancy':'Country'}, axis=1, inplace = True)
df_life = df_life.melt(id_vars="Country", var_name="Year", value_name="Life_expectancy" )
df_life.head(3)
```

```python
#Set the Indexes to Country
df_life.set_index('Country')
df_fer.set_index('Country')
df_pop.set_index('Country')
```

```python
#Convert the Datatype of the Year column in all dataframes to Integer. This is useful to do the merge operation
df_life['Year']= df_life['Year'].astype(int)
df_fer['Year']= df_fer['Year'].astype(int)
df_pop['Year']= df_pop['Year'].astype(int)
```

## Exploratory Data Analysis

```python
#Draw histograms for Life expectancy for Year 2015
df_life[df_life['Year']== 2015]['Life_expectancy'].hist(bins=20, facecolor='#ff0000', alpha=0.75, histtype='bar',)
plt.title('Life Expectancy 2015')
plt.xlabel('Life Expectancy')
plt.ylabel('Count')
plt.savefig('histo.png', dpi = 150)
```

```python
#Merge the three Dataframes on Country and Year
result = pd.merge(df_fer, df_life, how = 'left', on=['Country','Year'])
df_concat = pd.merge(result, df_pop, how = 'left', on=['Country','Year'])
df_concat
```

```python
#Remove the empty Cells
df_concat.dropna()
df_concat.isna().count()
```

```python
df_concat.head()
```

# Create a GIF with Countries as Labels

```python
#Create the plots for each Year
import seaborn as sns
import matplotlib.pyplot as plt
for year in range(1960, 2016):
    df_subset = df_concat[df_concat['Year'] == year]
    
    sns.scatterplot(x='Life_expectancy', y='Fertility', size = 'population' , data=df_subset, alpha=1).set(title = f'Year {year}')
    plt.xlabel('Life Expectancy in Years')
    plt.ylabel('Fertility Rate')
    
    plt.savefig('lifeexp_'+str(year)+'.png')
    plt.close()

    
```

```python
#Create a GIF File
import imageio

images = []

for i in range(1960, 2016):
    filename = 'lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output.gif', images, fps=5)
```

```python
#Display the GIF File
import ipywidgets as widgets
display(widgets.HTML(f'<img src="output.gif" width="750" align="center">'))
```

```python
#Continent dataframe preprocessing
df_con.rename(columns={'country':'Country'},inplace = True)
df_con.set_index('Country')
```

```python
df_concat.head()
```

```python
#Merge the Final Dataframe with the continents data
result = pd.merge(df_concat, df_con, how = 'left', on='Country')
result.head()
```

# Create a GIF with Continents as Labels


```python


#Create the plots for each Year
import seaborn as sns
import matplotlib.pyplot as plt
for year in range(1960, 2016):
    result_subset = result[result['Year'] == year]
    
    sns.scatterplot(x='Life_expectancy', y='Fertility', size = 'population' , data=result_subset, hue = 'continent', alpha=1).set(title = f'Year {year}')
    plt.xlim([20,150])
    plt.xlabel('Life Expectancy in Years')
    plt.ylabel('Fertility Rate')
    plt.savefig('lifeexpcon_'+str(year)+'.png')
    plt.close()
```

```python
#Create a GIF File
import imageio

images = []

for i in range(1960, 2016):
    filename = 'lifeexpcon_{}.png'.format(i)
    images.append(imageio.imread(filename))

imageio.mimsave('output1.gif', images, fps=5)
```

```python
#Display the GIF File
import ipywidgets as widgets
display(widgets.HTML(f'<img src="output1.gif" width="750" align="center">'))
```
